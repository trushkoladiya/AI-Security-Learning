## 1. What Python Basics Are

Python basics are the **fundamental building blocks** of Python programming. Think of them as the basic vocabulary and grammar rules you need to communicate with a computer using Python.

These basics include:
- **Keywords**: Special words Python reserves for its own use
- **Comments**: Notes you write for yourself (Python ignores them)
- **Indentation**: Spaces/tabs that show Python how code is organized
- **Print statements**: How you tell Python to show information on screen
- **Input handling**: How you ask users to give information to your program

Just like learning a new language requires knowing basic words and sentence structure, learning Python requires understanding these fundamental elements first.

## 3. How Python Basics Work

### Keywords
Keywords are special words that Python has already reserved for specific purposes. You **cannot** use them as variable names.

Examples: `if`, `else`, `for`, `while`, `def`, `return`, `True`, `False`, `None`.

Think of keywords like traffic signs—they have specific meanings you cannot change.

### Comments
Comments are lines in your code that Python ignores. They exist only for humans to read.

**Why use comments?**
- To explain what your code does
- To remind yourself why you wrote something a certain way
- To help others understand your code
- To temporarily disable code without deleting it

**How to write comments:**

````python
# This is a single-line comment
# Python ignores everything after the # symbol

x = 5  # You can also write comments after code
print(x)
````

### Indentation
Indentation means spaces or tabs at the beginning of lines. In Python, indentation is **NOT just for looks**—it actually tells Python which code belongs together.

Most languages use `{}` brackets to group code. Python uses indentation instead.

````python
if True:
    print("This is indented")
    print("This is also indented")

print("This is not indented")
````

### Print Statements
Print statements display information on the screen. This is how you see what your program is doing.

````python
print("Hello, World!")
print(42)
print("The answer is:", 42)
````

### Input Handling
Input lets your program ask the user for information and wait for them to type something.

````python
name = input("What is your name? ")
print("Hello", name)
````

## 5. Common Beginner Mistakes

### Mistake 1: Using keywords as variable names

Incorrect:
```
for = 5
if = "hello"
```

Correct:
```
count = 5
message = "hello"
```

### Mistake 2: Inconsistent indentation

Incorrect:
```
if True:
    print("First line")
  print("Second line")
```

Correct:
```
if True:
    print("First line")
    print("Second line")
```

### Mistake 3: Forgetting quotes in print statements

Incorrect:
```
print(Hello)
```

Correct:
```
print("Hello")
```

### Mistake 4: Not storing input result

Incorrect:
```
input("What is your name?")
print(name)
```

Correct:
```
name = input("What is your name?")
print(name)
```

### Mistake 5: Mixing tabs and spaces for indentation

Pro tip: Most professional programmers use **4 spaces** instead of tabs.
